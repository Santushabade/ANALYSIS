# Q5. Find dealers those are having RERA approved and that flat size 2 BHK then decrease price by 2 lakh & above 2 BHK the increase price by 7 lakh ?

### For  2 BHK decrease 2 in target price

```python
j=data.loc[(data["POSTED_BY"]=="Dealer")&(data["RERA"]==1)&(data["BHK_NO."]==2)]
j
```

```python
price2=[]
for x in j["TARGET(PRICE_IN_LACS)"]:
        k=x-2
        price2.append(k)
print(price2)
```

```python
j.insert(11,"NT",price2)
```

```python
j
```

## for above 2 BHK Increase 7 in Target price

```python
l=data.loc[(data["POSTED_BY"]=="Dealer")&(data["RERA"]==1)&(data["BHK_NO."]>2)]
l
```

```python
l["BHK_NO."].unique()
```

```python
price7=[]
for x in l["TARGET(PRICE_IN_LACS)"]:
    m=x+7
    price7.append(m)
print(price7)    
```

```python
l.insert(11,"NT2",price7)
l
```
